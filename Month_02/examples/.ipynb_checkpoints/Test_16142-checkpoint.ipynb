```python
Cop = cv.FDCT2D(smoothed_img.shape)

coeffs = Cop @ smoothed_img.ravel()
threshold_factor = 4.0  # needs_reconsideration

coeffs_thr = coeffs.copy()
sigma = np.std(coeffs)
coeffs_thr[np.abs(coeffs) < threshold_factor * sigma] = 0.0
reconstructed = np.abs((Cop.H @ coeffs_thr).reshape(smoothed_img.shape))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy.ndimage import gaussian_filter, sobel, binary_dilation, binary_erosion
from matplotlib.colors import LogNorm
import curvelops as cv
import warnings
warnings.filterwarnings('ignore')

# Load data
fits_file = "/home/rabbit/CASSA_Internship/Month_01/16142/repro/work/img/16142b4_0.5-2_flux.img"
with fits.open(fits_file) as hdul:
    img_data = hdul[0].data
    if img_data.ndim > 2:
        img_data = img_data[0]

print(f"Image shape: {img_data.shape}")
print(f"Min/max: {img_data.min():.2e}/{img_data.max():.2e}")

# ==================== PREPROCESSING ====================
# 1. Handle zeros/negatives
img_data_pos = np.where(img_data <= 0, np.percentile(img_data[img_data > 0], 5), img_data)

# 2. Variance stabilization for Poisson noise
img_stabilized = 2 * np.sqrt(img_data_pos + 3/8)

# 3. Light smoothing
smoothed_img = gaussian_filter(img_stabilized, sigma=0.5)

# ==================== CURVELET TRANSFORM ====================
# Initialize curvelet transform with available parameters
try:
    # Try different parameter names
    Cop = cv.FDCT2D(smoothed_img.shape, nbscales=6)
except TypeError:
    try:
        Cop = cv.FDCT2D(smoothed_img.shape, nscales=6)
    except:
        Cop = cv.FDCT2D(smoothed_img.shape)

print(f"Curvelet transform initialized: {Cop.shape} -> {Cop.dims}")

# Forward transform
coeffs_flat = Cop @ smoothed_img.ravel()

# ==================== ACCESS COEFFICIENTS BY SCALE ====================
# Get the structure of coefficients
coeffs_struct = Cop.struct(coeffs_flat)

print(f"\nNumber of scales: {len(coeffs_struct)}")

# ==================== SCALE-DEPENDENT THRESHOLDING ====================
# Parameters optimized for Chandra
# Start with simpler values
lambda_values = {}
for i in range(len(coeffs_struct)):
    if i == 0:  # Finest
        lambda_values[i] = 4.0
    elif i == len(coeffs_struct) - 1:  # Coarsest
        lambda_values[i] = 1.0
    else:  # Intermediate
        lambda_values[i] = 3.0 - (i / (len(coeffs_struct)-1)) * 2.0

print(f"Lambda values: {lambda_values}")

coeffs_thr_flat = np.zeros_like(coeffs_flat, dtype=np.complex128)
start_idx = 0

for scale_idx in range(len(coeffs_struct)):
    scale_coeffs = coeffs_struct[scale_idx]
    
    # Extract all coefficients for this scale
    if isinstance(scale_coeffs, list):
        # Multiple angles at this scale
        scale_data = []
        for angle_idx in range(len(scale_coeffs)):
            if scale_coeffs[angle_idx] is not None:
                scale_data.append(scale_coeffs[angle_idx].ravel())
        if scale_data:
            all_scale_coeffs = np.concatenate(scale_data)
        else:
            all_scale_coeffs = np.array([])
    else:
        # Coarsest scale (single array)
        all_scale_coeffs = scale_coeffs.ravel()
    
    if len(all_scale_coeffs) == 0:
        continue
    
    # Get real part for threshold calculation (curvelets are complex)
    all_scale_coeffs_real = np.real(all_scale_coeffs)
    
    # Robust noise estimation per scale (MAD on real part)
    med = np.median(all_scale_coeffs_real)
    mad = np.median(np.abs(all_scale_coeffs_real - med))
    sigma_scale = mad / 0.6745 if mad > 0 else np.std(all_scale_coeffs_real)
    
    # Apply scale-adaptive threshold
    threshold = lambda_values.get(scale_idx, 2.5) * sigma_scale
    
    # Apply threshold to this scale's coefficients
    end_idx = start_idx + len(all_scale_coeffs)
    coeffs_slice = coeffs_flat[start_idx:end_idx]
    
    # Threshold based on magnitude (absolute value of complex numbers)
    coeffs_magnitude = np.abs(coeffs_slice)
    coeffs_thr_flat[start_idx:end_idx] = coeffs_slice * (coeffs_magnitude > threshold)
    
    print(f"Scale {scale_idx}: ={lambda_values.get(scale_idx, 2.5):.1f}, "
          f"={sigma_scale:.4f}, threshold={threshold:.4f}, "
          f"kept {np.sum(coeffs_magnitude > threshold)/len(coeffs_slice)*100:.1f}%")
    
    start_idx = end_idx

# ==================== RECONSTRUCTION ====================
# Inverse transform
reconstructed_stab = (Cop.H @ coeffs_thr_flat).reshape(smoothed_img.shape)

# Take real part (should be almost entirely real anyway)
reconstructed_stab_real = np.real(reconstructed_stab)

# Inverse Anscombe transform (approximate)
reconstructed = (reconstructed_stab_real / 2)**2 - 3/8
reconstructed = np.where(reconstructed < 0, 0, reconstructed)

# ==================== EDGE DETECTION ====================
def simple_edge_detection(image):
    """Simple edge detection using gradient thresholding"""
    # Ensure we're working with real numbers
    image_real = np.real(image) if np.iscomplexobj(image) else image
    
    # Compute gradient magnitude
    grad_x = np.gradient(image_real, axis=0)
    grad_y = np.gradient(image_real, axis=1)
    gradient_magnitude = np.sqrt(grad_x**2 + grad_y**2)
    
    # Normalize
    grad_norm = (gradient_magnitude - gradient_magnitude.min()) / \
                (gradient_magnitude.max() - gradient_magnitude.min() + 1e-10)
    
    # Adaptive threshold (percentile-based) - ensure real numbers
    grad_norm_real = np.real(grad_norm) if np.iscomplexobj(grad_norm) else grad_norm
    threshold = np.percentile(grad_norm_real, 90)  # Top 10% as edges
    edges = grad_norm_real > threshold
    
    return edges, grad_norm_real

# Detect edges
edges, gradient = simple_edge_detection(reconstructed)

# Clean edges (optional)
edges_clean = binary_dilation(binary_erosion(edges, iterations=1), iterations=1)

# ==================== VISUALIZATION ====================
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Original image
vmin_orig = np.percentile(img_data[img_data > 0], 1)
vmax_orig = np.percentile(img_data, 99.5)
im0 = axes[0,0].imshow(img_data, origin='lower', cmap='inferno', 
                      norm=LogNorm(vmin=vmin_orig, vmax=vmax_orig))
axes[0,0].set_title('Original Chandra Image')
plt.colorbar(im0, ax=axes[0,0], fraction=0.046, pad=0.04)

# Variance stabilized
im1 = axes[0,1].imshow(smoothed_img, origin='lower', cmap='inferno')
axes[0,1].set_title('Variance Stabilized + Smoothed')
plt.colorbar(im1, ax=axes[0,1], fraction=0.046, pad=0.04)

# Curvelet reconstructed edges
vmax_rec = np.percentile(reconstructed, 99)
im2 = axes[0,2].imshow(reconstructed, origin='lower', cmap='inferno',
                       vmax=vmax_rec)
axes[0,2].set_title('Curvelet Edge Enhancement')
plt.colorbar(im2, ax=axes[0,2], fraction=0.046, pad=0.04)

# Original with edges overlay
im3 = axes[1,0].imshow(img_data, origin='lower', cmap='inferno',
                      norm=LogNorm(vmin=vmin_orig, vmax=vmax_orig))
# Plot edges as scatter points
y, x = np.where(edges_clean)
if len(x) > 0 and len(y) > 0:
    # Subsample if too many points
    if len(x) > 10000:
        step = len(x) // 10000
        x, y = x[::step], y[::step]
    axes[1,0].scatter(x, y, c='cyan', s=1, alpha=0.5, marker='.', 
                     edgecolors='none')
axes[1,0].set_title('Detected Edges Overlay')
plt.colorbar(im3, ax=axes[1,0], fraction=0.046, pad=0.04)

# Edge map
im4 = axes[1,1].imshow(edges_clean, origin='lower', cmap='gray')
axes[1,1].set_title('Binary Edge Map')
plt.colorbar(im4, ax=axes[1,1], fraction=0.046, pad=0.04)

# Gradient magnitude
im5 = axes[1,2].imshow(gradient, origin='lower', cmap='hot',
                      vmax=np.percentile(gradient, 99))
axes[1,2].set_title('Gradient Magnitude')
plt.colorbar(im5, ax=axes[1,2], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()

# ==================== STATISTICS ====================
print(f"\n=== STATISTICS ===")
print(f"Original image: min={img_data.min():.2e}, max={img_data.max():.2e}")
print(f"Reconstructed: min={reconstructed.min():.2e}, max={reconstructed.max():.2e}")
nonzero_coeffs = np.sum(np.abs(coeffs_thr_flat) > 0)
print(f"Nonzero coefficients kept: {nonzero_coeffs}/{len(coeffs_thr_flat)} "
      f"({nonzero_coeffs/len(coeffs_thr_flat)*100:.2f}%)")
print(f"Edge pixels detected: {edges_clean.sum()} "
      f"({edges_clean.sum()/edges_clean.size*100:.2f}% of image)")

# ==================== ENHANCED VISUALIZATION ====================
# Show just the edges more clearly
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original with contour
ax1 = axes[0]
ax1.imshow(img_data, origin='lower', cmap='inferno', 
           norm=LogNorm(vmin=vmin_orig, vmax=vmax_orig))
# Create contour from edges
contour_level = 0.5
ax1.contour(edges_clean, levels=[contour_level], colors='cyan', 
           linewidths=1.5, alpha=0.8)
ax1.set_title('Cold Front Candidates (Cyan Contours)')
ax1.set_xlabel('X (pixels)')
ax1.set_ylabel('Y (pixels)')

# Zoom on a region if you know where cold front might be
ax2 = axes[1]
# Use central region if you don't know specific coordinates
height, width = img_data.shape
y_center, x_center = height // 2, width // 2
size = min(height, width) // 3

y_slice = slice(max(0, y_center - size), min(height, y_center + size))
x_slice = slice(max(0, x_center - size), min(width, x_center + size))

ax2.imshow(img_data[y_slice, x_slice], origin='lower', cmap='inferno',
           norm=LogNorm(vmin=vmin_orig, vmax=vmax_orig))
# Plot edges in zoomed region
edges_zoom = edges_clean[y_slice, x_slice]
y_edges, x_edges = np.where(edges_zoom)
if len(x_edges) > 0:
    ax2.scatter(x_edges, y_edges, c='cyan', s=2, alpha=0.7, marker='.',
               edgecolors='none')
ax2.set_title('Zoomed Region with Edges')
ax2.set_xlabel('X (pixels)')
ax2.set_ylabel('Y (pixels)')

plt.tight_layout()
plt.show()

print("\n=== NEXT STEPS ===")
print("1. Adjust lambda_values if edges are too sparse or too noisy")
print("2. Try different nbscales values (4-8)")
print("3. Consider applying to unsharp-masked image for better contrast")
print("4. For cold fronts, look for curved, continuous edge structures")
```
